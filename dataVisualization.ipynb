```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# # load data from seaborn #
iris = sns.load_dataset('iris')

# # correlation among all features #
correlation = iris.corr(numeric_only=True)
correlation
```

```python

# # Shows correlations between numerical variables #
sns.heatmap(correlation, cmap = 'coolwarm',  annot=True),
sns.set_theme(style="whitegrid", color_codes=True) 
```

```python

# # Plots scatter + regression lines by group #
sns.lmplot(x="sepal_length", y="petal_length", hue="species", data = iris, palette="Set1", markers=["o", "s", "D"])

# # Plots scatter without best fit line #
sns.lmplot(x="sepal_length", y="petal_length", hue="species", data = iris, palette="Set1", fit_reg=False, markers=["o", "s", "D"])
plt.show()
```

```python

# # creates a grid of scatter plots for each pair of selected variables (columns), plus histograms or KDE plots on the diagonals #
sns.pairplot(iris, height=5, vars=["sepal_length", "petal_length"], markers=["o", "s", "D"], hue="species")

# # all parameter #
sns.pairplot(iris, height=5, markers=["o", "s", "D"], hue="species")
sns.pairplot(iris, height=5, kind="reg", hue="species", markers=["o", "s", "D"]) # with best fit line
plt.show()
```

```python

# # box for data Statistics or description #
sns.boxplot(data = iris, orient = "h")
plt.show()
```

```python

# # Basic value scatter, but overlapping  (stripplot for Categorical plot and scatterplot for Numeric vs numeric) #
sns.stripplot(x = "species", y = "sepal_length", hue="species", data = iris)
plt.show()
# # Better visibility, clearer distribution #
sns.stripplot(x = "species", y = "petal_length", hue="species", data = iris, jitter = True)
plt.figure(figsize=(10, 6))
# # Bee-swarm layout adjusts points to prevent overlap, Better than stripplot for small datasets (Best clarity, avoids overlap automatically) #
sns.swarmplot(x="species", y="petal_length", hue="species", data=iris)
plt.show()
```
