```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
import numpy as np
from sklearn import metrics
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

iris = load_iris()

knn_1 = KNeighborsClassifier(n_neighbors=1)
knn_5 = KNeighborsClassifier(n_neighbors=5)

X = iris.data  # # feature matrix #
y = iris.target # # response vector #

# # fit data #
knn_1.fit(X, y)
knn_5.fit(X, y)
```

```python
print(y) # y_true, means real answer #

y_predict1 = knn_1.predict(X) # model predicted answer #
print(y_predict1)

print('n_neighbors = 1 accuracy via knn :', knn_1.score(X,y)) # model accuracy via knn #

print('n_neighbors = 1 accuracy via metrics:', metrics.accuracy_score(y, y_pred = y_predict1)) # model accuracy via metrics #


y_predict5 = knn_5.predict(X)
print('n_neighbors = 5 accuracy via metrics:', metrics.accuracy_score(y, y_pred = y_predict5))
```

```python
from sklearn.metrics import confusion_matrix

# # how well a classification model performed, by comparing the actual values with the predicted values #
cm1 = confusion_matrix(y_true = y, y_pred=y_predict1)
cm5 = confusion_matrix(y_true = y, y_pred=y_predict5)

print(cm1)
print(cm5)

sns.heatmap(cm1, annot=True, cmap='Blues', fmt='d', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix for n_neighbors = 1')
plt.show()

sns.heatmap(cm5, annot=True, cmap='Blues', fmt='d', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix for n_neighbors = 5')
plt.show()
```
