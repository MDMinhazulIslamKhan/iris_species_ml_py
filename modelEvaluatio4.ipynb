```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

iris = load_iris()

X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 4, test_size = 0.4)

neighbors_settings = list(range(1,26))
scores = []

for k in neighbors_settings:
    knn  = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_predict = knn.predict(X_test)
    scores.append(metrics.accuracy_score(y_test, y_pred = y_predict))

scores
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(10, 6))
plt.plot(neighbors_settings, scores, marker='o', linestyle='-', color='blue', label='Accuracy')
plt.xlabel("Number of Neighbors (k)")
plt.ylabel("Testing Accuracy")
plt.title("KNN Accuracy vs k")
plt.grid(True) 
plt.xticks(neighbors_settings)
plt.legend()
plt.show()
```

```python
# # from 6 t0 18, accuracy is good #
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print(metrics.accuracy_score(y_test, y_pred))
```

```python
training_accuracy = []
test_accuracy = []

for k in neighbors_settings:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    training_accuracy.append(knn.score(X_train, y_train))
    test_accuracy.append(knn.score(X_test, y_test))


plt.plot(neighbors_settings, training_accuracy, label="Training Accuracy")
plt.plot(neighbors_settings, test_accuracy, label="Test Accuracy")
plt.xlabel("Number of Neighbors (k)")
plt.ylabel("Accuracy")
plt.title("Training vs Test Accuracy for KNN")
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(neighbors_settings, training_accuracy, label="Training Accuracy", marker='o')
plt.plot(neighbors_settings, test_accuracy, label="Test Accuracy", marker='s')
plt.xlabel("Number of Neighbors (k)")
plt.ylabel("Accuracy")
plt.title("Training vs Test Accuracy for KNN")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
